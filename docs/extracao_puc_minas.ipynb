```python
import pandas as pd
import glob
import os
```

```python
#Pega o working directory
downloads_path = os.path.join(os.path.expanduser("~"), "Downloads", "pucminas")

# Define o diretório de trabalho
os.chdir(downloads_path)
```

```python
# Lista todos os arquivos CSV dentro da pasta
arquivos_csv = glob.glob("*.csv")

dataframes = []

for i, arquivo in enumerate(arquivos_csv, start=1):
    nome_empresa = f'Empresa_{i:03d}'
    
    # Lê o CSV
    df = pd.read_csv(arquivo)
    
    # Substitui a coluna 'Projeto' pelo nome anonimizado
    if 'Projeto' in df.columns:
        df['Projeto'] = nome_empresa
    else:
        df['Projeto'] = nome_empresa
    
    # Força colunas 3 e 4 a valores fixos (Cliente / Analista)
    if df.shape[1] >= 4:  # garante que existam pelo menos 4 colunas
        df.iloc[:, 2] = "Cliente"
        df.iloc[:, 3] = "Analista"
    
    dataframes.append(df)
```

```python
# Junta todos os DataFrames
df_final = pd.concat(dataframes, ignore_index=True)
```

```python
# Salva o resultado na mesma pasta
df_final.to_csv("dados_anonimizados.csv", index=False)
```
